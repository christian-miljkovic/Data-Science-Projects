```python
import MySQLdb as mdb
import sys

con = mdb.connect(host = 'localhost', 
                  user = 'root', 
                  passwd = 'dwdstudent2015', 
                  charset='utf8', use_unicode=True);


# Create the Ebola database
db_name = 'Ebola'
create_db_query = "CREATE DATABASE IF NOT EXISTS {db} DEFAULT CHARACTER SET 'utf8'".format(db=db_name)

# Create a database
cursor = con.cursor()
cursor.execute(create_db_query)
cursor.close()
```

```python
cursor = con.cursor()
db_name = 'Ebola'
table_name = 'Cases'
# Create a table

create_table_query = '''CREATE TABLE IF NOT EXISTS {db}.{table} 
                                (cases int,
                                date datetime,
                                deaths int,
                                status varchar(250), 
                                PRIMARY KEY(date)
                                )'''.format(db=db_name, table=table_name)
cursor.execute(create_table_query)
cursor.close()
```

```python
from datetime import date, datetime, timedelta

query_template = '''INSERT INTO 
    Ebola.Cases(cases, date, deaths, status) 
    VALUES (%s, %s, %s, %s)'''

cursor = con.cursor()

for entry in data:
    cases = entry['cases']
    
    date_split = entry['date'][0:10]
    
    date = datetime.strptime(date_split, 
                         '%Y-%m-%d')

    deaths = entry['deaths']
    status = entry['status']
    
    
    print("Inserting case", cases)
    query_parameters = (cases, date, deaths, status)
    cursor.execute(query_template, query_parameters)

con.commit()
cursor.close()
```

```python
cur = con.cursor(mdb.cursors.DictCursor)
cur.execute("SELECT * FROM Ebola.Cases")
rows = cur.fetchall()
```

```python
for row in rows:
    print("Case ID:", row["cases"])
    print("Date:", row["date"])
    print("Number of Deaths:", row["deaths"])
    print("Status:", row["status"])
    print("=============================================")
    
cur.close()
```
